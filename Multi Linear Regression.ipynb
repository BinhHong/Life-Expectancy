# 1. Use matrix

```python
data.columns
```

```python
# remove some columns from the first step and add columns 1's
modified_data = data.assign(all_one = np.ones(len(data)))
```

```python
# train and test sets

train = modified_data.iloc[:int(0.7*len(data)),:]
test = modified_data.iloc[int(0.7*len(data)):,:]
```

```python
train
```

```python
test
```

```python
# take features and target
x_train = train.drop(['Year', 'Life Expectancy'], axis=1)

y_train = train['Life Expectancy']

x_test = test.drop(['Year', 'Life Expectancy'], axis=1)

y_test = test['Life Expectancy']
```

```python
x_train
```

```python
# features matrix
A = x_train.values

# target matrix
y = y_train.values
```

```python
# sequence (b1,..., b18, b0)
B = np.dot(A.T,A)
C = np.dot(np.linalg.inv(B),np.dot(A.T,y))
```

```python
# this is the sequence of parameters
C
```

```python
# how to compute predictions?

# matrix of test set
B = x_test.values
```

```python
# predictions
y_prediction = np.dot(B, C)
```

```python
y_prediction
```

```python
y_test
```

```python
d = {'y_prediction': y_prediction, 'y_test': y_test.values}
```

```python
pd.DataFrame(data=d)
```
